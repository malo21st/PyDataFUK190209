# 入力（データの読込）　import pandas as pd
- **pd.read_excel()**　pd.read_csv()　pd.read_json()　pd.read_html()

```python
import pandas as pd

#URL = 'https://ckan.open-governmentdata.org/dataset/90a2e0fd-e466-4526-8e8d-c3d7e1293b82/resource/2abde841-8d4b-468e-8ed5-bd51d6967076/download/972.xls'
#df = pd.read_excel(URL)

df = pd.read_excel('972.xls')
```

```python
print(df.shape)
df.head(20)
```

---

# 入力（前処理）　df  DataFrame
- **df.iloc[ ]　df.T**　df.loc[ ]　df.[df[0].str.contains("〜")==True]


```python
# python2　　type(range(10)) => <type 'list'>
# python3　　type(range(10)) => <type 'range'>

raw = list(range(11,70,6)) + list(range(72,134,6))  # [11, 17, ..., 65, 72, 78, ..., 132]
col = list(range(2,36,3))                           # [2, 5, 8, 11, ..., 32, 35 ] 
df_male_temp = df.iloc[raw, col]

print(df_male_temp.shape)
df_male_temp        # 行：年代別の人口　，列：年
```

```python
df_male = df_male_temp.T

print(df_male.shape)
df_male.tail()        # 行：年　，列：年代別の人口
```

```python
raw = list(range(11,70,6)) + list(range(72,134,6))
col = list(range(3,37,3))
df_female_temp = df.iloc[raw, col]
df_female = df_female_temp.T

type(df_female)
```

---

# 処理　import numpy as np
- **n_ary = np.array(df.values)　n_ary1 + nary2　n_ary.sum(axis=int)**


```python
import numpy as np

# pandasu.df => numpy.array
np_male   = np.array(df_male.values)
np_female = np.array(df_female.values)

type(np_female)
```

```python
# 男女の5歳階級別人口を合計する
np_total = np_male + np_female

print(np_total[3, 5], np_male[3, 5], np_female[3, 5])
```

```python
# 年ごと（1955年〜2010年）の総人口を求める
np_yearly = np_total.sum(axis=1)

np_yearly

```

```python
# ６５歳以上の高齢者人口
np_old_yearly = np_total[:, 13:].sum(axis=1)

np_old_yearly
```

```python
# 高齢化率　65歳以上が占める割合
np_old_rate = np_old_yearly / np_yearly * 100

np_old_rate
```

# 出力　import matplotlib as plt   / gridspec
- **gridspec.GridSpec()　ax = plt.subplot()　ax.plot()**

- 人口ピラミッド　出典「**みんなのPython 第四版**」サンプルコード
<img src="img/minpy4th.jpg">

- **Chapter 12-04　日本の人口を視覚化する**　　人口ピラミッドを描く
 - https://coreblog.org/ats/stuff/minpy_support/samplecodes04/samplecodes/Chapter12/12-04.html


```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import gridspec

def show_pgraph(year, arr1, arr2, arr3,
                ymin, ymax, ydim=1):
    # 人口ピラミッドを表示する
    # 表示する人口のインデックスを得る
    idx = int((year-ymin)/ydim)
    # 人口ピラミッドと人口グラフのグリッドを生成
    gs = gridspec.GridSpec(2, 2, height_ratios=(3, 2))
    # グラフの配置を決める
    ax = [plt.subplot(gs[0, 0]),plt.subplot(gs[0, 1]),
          plt.subplot(gs[1, :])]
    # 男性の人口ピラミッドを描く
    ax[0].barh(range(0, 101, 5), arr1[idx], height=3)
    ax[0].set(ylim=(0, 105), xlim=(0, 300000))
    ax[0].invert_xaxis()
    ax[0].yaxis.tick_right()
    # 女性の人口ピラミッドを描く
    ax[1].barh(range(0, 101, 5), arr2[idx], height=3)
    ax[1].tick_params(labelleft=False)                   #  'off'  =>  False
    ax[1].set(ylim=(0, 105), xlim=(0, 300000))
    # 人口グラフを描き西暦に線を引く
    ax[2].plot(range(ymin, ymax+1, ydim), arr3, ls=":")
    ax[2].plot([year, year], [0, 5500000])
    
```

```python
# show_pgraph(year, arr1, arr2, arr3, arr4, ymin, ymax, ydim=1)
show_pgraph(1955, np_male, np_female, np_yearly, 1955, 2010, 5)
```

# 出力　ipywidgets.interact, IntSlider, fixed
- **interact(func, year=IntSlider( ), arr1=fixed(df), .....)**



```python
# スライダを使って人口ピラミッドを動かす
from ipywidgets import interact, IntSlider, fixed

print("福岡県の人口ピラミッド　１９５５〜２０１０年")
t = interact(show_pgraph,
             year=IntSlider(min=1955, max=2010, step=5, description='西暦'),
             arr1=fixed(np_male),
             arr2=fixed(np_female),
             arr3=fixed(np_yearly),
             ymin=fixed(1955),
             ymax=fixed(2010), 
             ydim=fixed(5)
            )
```

## デコレータでも使えます。 @interact(  )

```python
# スライダを使って人口ピラミッドを動かす　　（デコレータ版）
from ipywidgets import interact, IntSlider, fixed

print("福岡県の人口ピラミッド　１９５５〜２０１０年")

selected_year=IntSlider(min=1955, max=2010, step=5, description='西暦')

@interact(year=selected_year, 
          arr1=fixed(np_male), 
          arr2=fixed(np_female), 
          arr3=fixed(np_yearly), 
          arr4=fixed(np_old_rate),
          ymin=fixed(1955), 
          ymax=fixed(2010), 
          ydim=fixed(5)
         )
def show_pgraph1(year, arr1, arr2, arr3, arr4, ymin, ymax, ydim=1):
    # 人口ピラミッドを表示する
    # 表示する人口のインデックスを得る
    idx = int((year-ymin)/ydim)
    # 人口ピラミッドと人口グラフのグリッドを生成
    gs = gridspec.GridSpec(2, 2, height_ratios=(3, 2)) # , width_ratios=(1, 2)
    # グラフの配置を決める
    ax = [plt.subplot(gs[0, 0]),plt.subplot(gs[0, 1]),
          plt.subplot(gs[1, :])]
    # 男性の人口ピラミッドを描く
    ax[0].barh(range(0, 101, 5), arr1[idx], height=5, color='b', alpha=0.5, label="男性")
    ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x)))) # 三桁カンマの設定
    ax[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, loc: "{:>3}".format(int(y)))) 
    ax[0].set(ylim=(0, 105), xlim=(0, 300000))
    ax[0].invert_xaxis()
    ax[0].yaxis.tick_right()
    ax[0].legend(loc='upper left')
    # 女性の人口ピラミッドを描く
    ax[1].barh(range(0, 101, 5), arr2[idx], height=5, color='r', alpha=0.5, label="女性")
    ax[1].tick_params(labelleft=False)
    ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x)))) # 三桁カンマの設定
    ax[1].set(ylim=(0, 105), xlim=(0, 300000))
    ax[1].legend(loc='upper right')
    # 人口グラフを描き西暦に線を引く
    ax[2].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, loc: "{:,}".format(int(y)))) # 三桁カンマの設定
    ax[2].set_ylim([0,5500000])
    ax[2].plot(range(ymin, ymax+1, ydim), arr3, ls="-", label="人口[人]", color="r", alpha=0.5)
    ax[2].legend(loc='center left')
    ax[2] = ax[2].twinx()
    ax[2].set_ylim([0,30])
    ax[2].plot(range(ymin, ymax+1, ydim), arr4, ls="-.", label="高齢化率[%]")
    ax[2].plot([year, year], [0, 5000000])
    ax[2].legend(loc='lower right')
    
```
